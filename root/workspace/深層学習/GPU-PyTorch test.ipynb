```python
import torch

# PyTorchのバージョンを表示
print(torch.__version__)

#PyTorchでGPUが使用可能か確認(Trueなら使用可能)
print(torch.cuda.is_available())

#GPUがいくつ使用可能か
print(torch.cuda.device_count())

# デフォルトのGPU番号（インデックス）
print(torch.cuda.current_device())
```
